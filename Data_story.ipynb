### What courses have the best reviews? 

```python
# create datafrmaes for each survey question response 
df_great = df_course[df_course['Student Survey Response']=='Great']
df_okay = df_course[df_course['Student Survey Response']=='Okay']
df_needs_improvement = df_course[df_course['Student Survey Response']=='Needs improvement']
df_no_opinion = df_course[df_course['Student Survey Response']=='No opinion']
df_review_all = df_course[(df_course['Student Survey Response']=='No opinion')|(df_course['Student Survey Response']=='Great')
                         |(df_course['Student Survey Response']=='Okay')|(df_course['Student Survey Response']=='Needs improvement')]
```

```python
# define function to calculate the percent of each response 
def percent_response(df):
    df_temp = df['Course Name'].value_counts().sort_index(ascending=False)
    df_temp2 = df_course['Course Name'].value_counts().sort_index(ascending=False)
    df_temp3 = df_temp/df_temp2.sort_values()
    df = pd.DataFrame(df_temp3)
    df = df.sort_values(by='Course Name',ascending=False)
    return df
```

```python
# Calculate percentage
df_great_percent = percent_response(df_great)
df_okay_percent = percent_response(df_okay)
df_needs_improvement_percent = percent_response(df_needs_improvement)
df_no_opinion_percent = percent_response(df_no_opinion)
```

```python
#reset Index
df_great_percent.reset_index(inplace=True)
df_okay_percent.reset_index(inplace=True)
df_needs_improvement_percent.reset_index(inplace=True) 
df_no_opinion_percent.reset_index(inplace=True) 
```
